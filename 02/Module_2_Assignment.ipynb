In this practical assignment, you will be using the "Wisconsin Breast Cancer" dataset. The dataset consists of 569 instances used for diagnosing Malignant (M) and Benign (B) cancer types. Your tasks are as follows:

### **1. Separate the variables that can be used and the ones that cannot be used.**

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

data = 'JS2-LW-Assignment-Wbc.csv' # path dataset
df = pd.read_csv(data) # load dataset
```

```python
dpath = 'JS2-LW-Assignment-Wbc.csv'
df = pd.read_csv(dpath)
df.head()
```

```python
df = df[['diagnosis','radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean','compactness_mean','concavity_mean', 'concave points_mean','texture_worst','perimeter_worst', 'area_worst', 'smoothness_worst', 'compactness_worst', 'concavity_worst', 'concave points_worst','symmetry_worst', 'fractal_dimension_worst'  ]]
df.head()
```

### **2. Perform encoding on the "diagnosis" column.**

```python
le = LabelEncoder() # initiate LabelEncoder object
df['diagnosis'] = le.fit_transform(df['diagnosis']) # encoding
```

```python
df.head()
```

### **3. Standardize all columns with numeric values.**

```python
std = StandardScaler()
df['diagnosis'] = std.fit_transform(df[['diagnosis']])
df['radius_mean'] = std.fit_transform(df[['radius_mean']])
df['texture_mean'] = std.fit_transform(df[['texture_mean']])
df['perimeter_mean'] = std.fit_transform(df[['perimeter_mean']])
df['area_mean'] = std.fit_transform(df[['area_mean']])
df['texture_worst'] = std.fit_transform(df[['texture_worst']])
df['perimeter_worst'] = std.fit_transform(df[['perimeter_worst']])
df['area_worst'] = std.fit_transform(df[['area_worst']])
```

```python
df.head()
```

### **4. Perform a stratified split of the data to create training and test data with an 80:20 ratio.**

```python
from sklearn.model_selection import train_test_split

# Split data into training (80%) and testing (20%)
df_train, df_test = train_test_split(df, test_size=0.2, random_state=0, stratify=df['diagnosis'])

# Check the amount of data in each subset
print(f'Amount of original data :\n{df.diagnosis.value_counts()}')
print()
print(f'Amount of training data :\n{df_train.diagnosis.value_counts()}')
print()
print(f'Number of testing data :\n{df_test.diagnosis.value_counts()}')
```
