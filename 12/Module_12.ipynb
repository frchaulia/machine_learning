## **Image Preparation**

```python
import cv2
from google.colab.patches import cv2_imshow
from google.colab import files

# Read image
image = cv2.imread('hitchhikers-rotated.png')

# Convert from BGR to RGB
b, g, r = cv2.split(image)
rgb_img = cv2.merge([r, g, b])

# Display image
cv2_imshow(rgb_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## **Orientation and Script Detection**

```python
!sudo apt install tesseract-ocr
!pip install pytesseract
```

```python
import pytesseract
from pytesseract import Output

# OCR
results = pytesseract.image_to_osd(rgb_img, output_type=Output.DICT)

# Display
print("Orientation: {}".format(results["orientation"]))
print("Rotation {} degrees to correct".format(results["rotate"]))
print("Script: {}".format(results["script"]))
```

```python
import matplotlib.pyplot as plt
import imutils

# Rotate the image
rotated = imutils.rotate_bound(rgb_img, angle=results["rotate"])

# Display rotated image
plt.figure(figsize=(8, 16))
plt.imshow(rotated, cmap='gray')
plt.title('Rotated Image Result')
plt.axis('off')
plt.show()
```

## **Perform OCR**

```python
custom_config = r'-l eng --oem 3 --psm 6'
result = pytesseract.image_to_string(rotated, config=custom_config)
print(result)
```
